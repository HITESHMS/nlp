```python
import nltk
nltk.download('punkt')
```

```python
from nltk.util import ngrams
```

```python
data ="All your dreams can come true if we have the courage to pursue them."
```

```python
tokens= nltk.word_tokenize(data) 
```

```python
tokens
tokenized_bigrams = list(nltk.bigrams(tokens))
tokenized_bigrams
```

```python
tokenized_trigrams = list(nltk.trigrams(tokens))
tokenized_trigrams
```

```python
tokenized_ngrams = list(nltk.ngrams(tokens,5))
tokenized_ngrams 
```
